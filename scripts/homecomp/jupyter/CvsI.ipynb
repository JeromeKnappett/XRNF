```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
cLx = pickle.load(open('/home/jerome/dev/scripts/jupyter/cXlength.pkl', 'rb'))
cLy = pickle.load(open('/home/jerome/dev/scripts/jupyter/cYlength.pkl', 'rb'))
Ix = pickle.load(open('/home/jerome/dev/scripts/jupyter/totIx.pkl', 'rb'))
Iy = pickle.load(open('/home/jerome/dev/scripts/jupyter/totIy.pkl', 'rb'))
```

```python
sX = [25,50,75,100,125,150,175,200,225,250,300,350]
sY = [25,50,75,100,125,150,175,200,225,250]
```

```python
print(np.shape(sX))
print(np.shape(Ix))

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(sX,Ix[0:12],'x:',color='red')#,label='intensity')
ax2.plot(sX,[c*1e3 for c in cLx[0:12]],'x:',color='blue')#,label='coherence length')
ax1.set_xlabel('Slit separation [microns]')
ax1.set_ylabel('Intensity')
ax2.set_ylabel('Coherence Length [mm]')
ax1.spines['left'].set_color('red')
ax1.spines['right'].set_color('blue')

ax1.yaxis.label.set_color('red')
ax1.tick_params(axis='y', colors='red')
ax2.yaxis.label.set_color('blue')
ax2.tick_params(axis='y', colors='blue')
ax1.set_title('Horizontal')
fig.tight_layout()
plt.show

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(sY,Iy[0:10],'x:',color='red')#,label='intensity')
ax2.plot(sY,[c*1e3 for c in cLy[0:10]],'x:',color='blue')#,label='coherence length')
ax1.set_xlabel('Slit separation [microns]')
ax1.set_ylabel('Intensity')
ax2.set_ylabel('Coherence Length [mm]')
ax1.spines['left'].set_color('red')
ax1.spines['right'].set_color('blue')

ax1.yaxis.label.set_color('red')
ax1.tick_params(axis='y', colors='red')
ax2.yaxis.label.set_color('blue')
ax2.tick_params(axis='y', colors='blue')
ax1.set_title('Vertical')
fig.tight_layout()
plt.show
```
