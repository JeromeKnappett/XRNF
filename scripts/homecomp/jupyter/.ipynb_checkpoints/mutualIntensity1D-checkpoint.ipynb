# Plotting the Mutual Intensity of a Complex Electric Field in Python

## Step One
Create a one dimensional Gaussian function to represent the complex electric field profile

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import imageio
```

First make a function to create a one dimensional gaussian, with the option of adding complex values.

```python
def gauss1D(n,R,A,sigma,show = True,complex=False):
    """
    n: number of points
    R: range of position values
    A: amplitude
    sigma: standard deviation
    show: T/F, turn plotting on or off
    complex: choose function to define imaginary part, can be 'sinc', 'cos', or 'rand'.
    """
    
    dx = np.linspace(-R/2,R/2,n)
    G = [A*np.exp((-1*(x**2))/(2*(sigma**2))) for x in dx] # Gaussian distribution over dx
    
    # add imaginary part to represent phase, either 'sinc', 'cos', or 'rand'
    if complex == 'sinc':
        C = (np.pi)*(np.divide(np.sin(dx),dx,np.ones_like(dx),where=dx!=0)) # add an arbitrary imaginary part (pi*sinc(x))
    elif complex == 'cos':
        C = np.pi*np.cos(-dx/2)*np.ones_like(dx)
    elif complex == 'rand':
        C = [c/100 for c in random.sample(range(-314, 314), n)]
    
    if complex:
        G = np.array([g + 1j*c for g,c in zip(G,C)])   # convert G into complex
        I = abs(G.conjugate()*G)**2
    else:
         pass
    
    if show:
        if complex:
            plt.plot(dx,np.real(G),'o', label='real part')
            plt.plot(dx,C, 'x', label='imaginary part')
            plt.plot(dx,I/np.max(I)*np.max(np.real(G)), '.', label='intensity') # intensity is normalised for better plotting
            plt.legend()
        else:
            plt.plot(dx,G, 'o')
        plt.xlabel('Position')
        plt.ylabel('Amplitude')
        plt.show()
    
    return G
```

```python
n = 101
R = n/8
A = 10
sigma = 1

G = gauss1D(n,R,A,sigma,show=True,complex='sinc')     # gaussian with imaginary part given by sinc function
#G = gauss1D(n,R,A*10,sigma,show=True,complex='rand') # gaussian with imaginary part given by random values
#G = gauss1D(n,R,A*2,sigma,show=True,complex='cos')   # gaussian with imaginary part given by cos function
```

## Step Two

Create a function to calculate the mutual intensity, J, from a one dimensional array of complex values.

```python
def testCarray1D(A,B):
    
    J = []
    C = []
    
    #1D approach (A & B are 1D)
    for x, a in enumerate(A):         # x = x1, a = E1
        for i, b in enumerate(B):     # i = x2, b = E2
            # define new variables
            _x = (x+i)/2              # x  = (x1+x2)/2
            dx = (x-i)/2              # dx = (x1-x2)/2
            
            # get mutual intensity
            j = a.conjugate()*b       # J = <E1*(x1)E2(x2)>
            
            # get degree of coherence   c = J/[I(x1)I(x2)]^(1/2)
            c = j/np.sqrt((a.conjugate()*a)*(b.conjugate()*b))
            
            # create array of j values for x, dx
            J.append((j,_x,dx))       # J(x,dx)
            C.append((c,_x,dx))       # C(x,dx)
            
            #print(f'J(x,dx): {j}')
            #print(f'(x,dx): {(_x, dx)}')
    
    print(np.shape(J))
    print(J[0])
    _j = [j[0] for j in J]
    _j = np.reshape(_j,(np.size(A),np.size(A)))
    _c = [c[0] for c in C]
    _c = np.reshape(_c,(np.size(A),np.size(A)))
    
    plt.imshow(abs(np.array(_j)),aspect='auto')
    plt.show()
    
    plt.plot(abs(_j[int(np.size(A)/2)]))
    plt.show()
    
    I = abs(_j[int(np.size(A)/2)])
    
    # summing J and C values
    Jsum = {}
    J = sorted(J, key=lambda tup: tup[2]) # sorting J values by ascending dx (for nice plots)
    Csum = {}
    C = sorted(C, key=lambda tup: tup[2]) # sorting C values by ascending dx (for nice plots)
    
    for j, x, dx in J:
        Jtot = Jsum.get(dx, 0) + j  # summing J(x,dx) for all unique dx
        Jsum[dx] = Jtot             # J(dx)
    for c, x, dx in C:
        Ctot = Csum.get(dx, 0) + c  # summing C(x,dx) for all unique dx
        Csum[dx] = Ctot             # C(dx)
    
    # separating J and dx
    DX = np.fromiter(Jsum.keys(), dtype=float)           # array of dx values
    jX = abs(np.fromiter(Jsum.values(), dtype=complex))  # array of J values
    cX = abs(np.fromiter(Csum.values(), dtype=complex))  # array of J values
    
    #print(DX)
    
    return jX, DX, I, cX
```

## Step Three

Calculate and plot the mutual intensity of the gaussian generated above

```python
J, dx, I, C = testCarray1D(G,G)

plt.plot(dx,J, '--')
plt.xlabel('dx')
plt.ylabel('J')
plt.show()

plt.plot(dx,C, '--')
plt.xlabel('dx')
plt.ylabel('C')
plt.show()
```

## Step Four

Normalise the mutual intensity by the squared intensity to get the spatial coherence.

As seen below, the squared intensity is much greater than the mutual intensity, so I am unsure if the results of this step are accurate.

```python
# intensity
#I = abs(np.array(G).conjugate()*np.array(G))
xp = np.linspace(-n/2,n/2,n)  # position values to plot intensity over (intensity array is ~half the size of J)

# print maximum values of squared intensity, intensity and mutual intensity
print(f'Max I^2: {np.max(I**2)}')
print(f'Max I:   {np.max(I)}')
print(f'Max J:   {np.max(J)}')

# plotting J and I^2
plt.plot(dx,J, 'x', label='J')
plt.plot(xp,I**2, 'o', label='$I^2$')
plt.xlabel('dx')
plt.ylabel('amplitude')
plt.legend()
plt.show()

# plotting gamma (spatial coherence)
plt.plot(dx,J/np.max(I**2), 'x', label='$\gamma$')
plt.xlabel('dx')
plt.ylabel('amplitude')
plt.legend()
plt.show()

# plotting normalised J and I
plt.plot(dx,J/np.max(J), 'x', label='normalised J')
plt.plot(xp,I/np.max(I), 'o', label='normalised I')
plt.xlabel('dx')
plt.ylabel('amplitude')
plt.legend()
plt.show()

# print the shape of J and I arrays
print(f'Shape of J: {np.shape(J)}')
print(f'Shape of I: {np.shape(I)}')
```

As shown in the previous cell, the mutual intensity array (J) is around twice the size of the intensity array (I). This means that J needs to be normalised by just the center value of the squared intensity.

This is another aspect that leads me to beleive I may be doing something wrong.

## Step five

Load a simulated electric field for analysis.

```python
# load complex electric field tiffs
EhR = imageio.imread('beforeBDA_efield_sx200sy200ExReal.tif')  # real part of horizontal component
EvR = imageio.imread('beforeBDA_efield_sx200sy200EyReal.tif')  # real part of vertical component
EhI = imageio.imread('beforeBDA_efield_sx200sy200ExIm.tif')    # imaginary part of horizontal component
EvI = imageio.imread('beforeBDA_efield_sx200sy200EyIm.tif')    # imaginary part of vertical component

#print(np.shape(EhR))
#print(np.shape(EvR))
#print(np.shape(EhI))
#print(np.shape(EvI))

# Combine real and imaginary parts of each component
Eh = EhR + EhI*1j # horizontal component of complex electric field  
Ev = EvR + EvI*1j # vertical component of complex electric field
cE = Ev + Eh      # complex electric field
```

```python
# plot intensity of simulated electric field
Isim = abs(cE.conjugate()*cE)

plt.imshow(Isim, aspect='auto')
plt.show()

print(np.shape(EvI))
```

```python
# define sampling range for profile
ne = 550 
Re = 2*ne

#resolution of electric field (from simulation)
dx200, dy200 = 2.0296997983770577e-06, 1.7427838524854241e-06

# get horizontal and vertical profile of electric field
ex = cE[int(np.shape(cE)[0]/2),int(np.shape(cE)[1]/2)-ne:int(np.shape(cE)[1]/2)+ne]
ey = cE[int(np.shape(cE)[0]/2)-ne:int(np.shape(cE)[0]/2)+ne,int(np.shape(cE)[1]/2)]

Jx, dxsim, Ix, Cx = testCarray1D(ex,ex)
Jy, dysim, Iy, Cy = testCarray1D(ey,ey)

plt.plot(dxsim*dx200*1e6,Jx, '--')
plt.xlabel('dx [um]')
plt.ylabel('Jx')
plt.show()

plt.plot(dysim*dy200*1e6,Jy, '--')
plt.xlabel('dy [um]')
plt.ylabel('Jy')
plt.show()

plt.plot(dxsim*dx200*1e6,Cx, '--')
plt.xlabel('dx [um]')
plt.ylabel('Jx')
plt.show()

plt.plot(dysim*dy200*1e6,Cy, '--')
plt.xlabel('dy [um]')
plt.ylabel('Jy')
plt.show()

# intensity
#Ix = abs(np.array(ex).conjugate()*np.array(ex))
#Iy = abs(np.array(ey).conjugate()*np.array(ey))
x = np.linspace(-ne*dx200*1e6,ne*dx200*1e6,Re)  
y = np.linspace(-ne*dy200*1e6,ne*dy200*1e6,Re)  

# print maximum values of squared intensity, intensity and mutual intensity
print(f'Max I^2: {np.max(Ix**2)}')
print(f'Max I:   {np.max(Ix)}')
print(f'Max J:   {np.max(Jx)}')

# plotting normalised J and I
plt.plot(dxsim*dx200*1e6,Jx/np.max(Jx), 'x', label='normalised J')
plt.plot(x,Ix/np.max(Ix), 'o', label='normalised I')
plt.xlabel('dx [um]')
plt.ylabel('amplitude')
plt.legend()
plt.show()

plt.plot(dysim*dy200*1e6,Jy/np.max(Jy), 'x', label='normalised J')
plt.plot(y,Iy/np.max(Iy), 'o', label='normalised I')
plt.xlabel('dx')
plt.ylabel('amplitude')
plt.legend()
plt.show()
```
